```python
import seaborn as sns
```

```python
#Define o tamanho da imagem.
sns.set(rc={'figure.figsize':(20,10)}) 
#Carrega os dados
dados = sns.load_dataset("planets")
#Plota os dados
grafico = sns.countplot(x="year", data=dados,color = 'lightblue')
#Renomeia os eixos
grafico.set(xlabel='Ano', ylabel='Numero de Planetas')
```
